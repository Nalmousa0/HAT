```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
from selenium.webdriver.common.keys import Keys
import time

# Initialize WebDriver
driver = webdriver.Chrome()


 # Switch to the main window/tab
# Initialize a list to store the data
data_list = []
# Open the webpage
driver.get("https://rasaif.com/?cat=85")

page_source = driver.page_source
soup = BeautifulSoup(page_source, "html.parser")

# Extract and process the data using BeautifulSoup  
sections = soup.find_all(class_="elementor-inner-section")
for section in sections:
    columns = section.find_all(class_="elementor-column")
    if len(columns) == 2:
        arabic_text = columns[0].find("p")
        english_translation = columns[1].find("p")
        if arabic_text and english_translation:
            arabic_text = arabic_text.get_text(strip=True)
            english_translation = english_translation.get_text(strip=True)
            data_list.append([arabic_text, english_translation])

# Define the number of times to click the "Load More" button
num_of_clicks = 10000

def wait_for_next_button():
    return WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.XPATH, "//div[@class='jet-filters-pagination__link' and contains(text(), 'التالي')]"))
        
    )
    
# Wait for page to fully load

while num_of_clicks > 0:
    try:
        next_button = wait_for_next_button()
        driver.execute_script("arguments[0].scrollIntoView();", next_button)
        actions = ActionChains(driver)
        
        actions.move_to_element(next_button).click().perform()
        time.sleep(3)  # Adjust the sleep time if needed
        num_of_clicks -= 1
    except Exception as e:
        print('Cannot click the next button', e)
        break
    
    WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "elementor-inner-section")))




# Get the page source and parse it using BeautifulSoup
    page_source = driver.page_source
    soup = BeautifulSoup(page_source, "html.parser")

    # Extract and process the data using BeautifulSoup  
    sections = soup.find_all(class_="elementor-inner-section")
    for section in sections:
        columns = section.find_all(class_="elementor-column")
        if len(columns) == 2:
            arabic_text = columns[0].find("p")
            english_translation = columns[1].find("p")
            if arabic_text and english_translation:
                arabic_text = arabic_text.get_text(strip=True)
                english_translation = english_translation.get_text(strip=True)
                data_list.append([arabic_text, english_translation])

# Print the stored data
for item in data_list:
    arabic_text, english_translation = item
    print("Arabic:", arabic_text)
    print("English:", english_translation)
    print("=" * 40)

# Close the WebDriver
driver.quit()
```
