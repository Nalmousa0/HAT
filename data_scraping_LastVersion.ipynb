```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time

# Initialize WebDriver
driver = webdriver.Chrome()

# Open the webpage
driver.get("https://rasaif.com/?cat=131")

# Define the number of times to click the "Load More" button
num_of_clicks = 1000

# Function to wait for the "Load More" button to be clickable
def wait_for_load_more_button():
    return WebDriverWait(driver, 10).until(
        EC.element_to_be_clickable((By.CLASS_NAME, "elementor-button-text"))
    )

# Loop to click the "Load More" button
for i in range(num_of_clicks):
    try :
        load_more_button = wait_for_load_more_button()
        actions = ActionChains(driver)
        actions.move_to_element(load_more_button).click().perform()
        time.sleep(2)  # Adjust the sleep time if needed
        print(i)
    except Exception as e :
        print ('Reach the last page' , e)
        break


# Wait for the page to load after all clicks
WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "elementor-inner-section")))

# Initialize a list to store the data
data_list = []

# Get the page source and parse it using BeautifulSoup
page_source = driver.page_source
soup = BeautifulSoup(page_source, "html.parser")

# Extract and process the data using BeautifulSoup
sections = soup.find_all(class_="elementor-inner-section")
for section in sections:
    columns = section.find_all(class_="elementor-column")
    if len(columns) == 2:
        arabic_text = columns[0].find("p")
        english_translation = columns[1].find("p")
        if arabic_text and english_translation:
            arabic_text = arabic_text.get_text(strip=True)
            english_translation = english_translation.get_text(strip=True)
            data_list.append([arabic_text, english_translation])

# Print the stored data
for item in data_list:
    arabic_text, english_translation = item
    print("Arabic:", arabic_text)
    print("English:", english_translation)
    print("=" * 40)

# Close the WebDriver
driver.quit()
```

```python
import pandas as pd
df = pd.DataFrame(data_list,columns = ['Arabic', 'English'])
df
```
